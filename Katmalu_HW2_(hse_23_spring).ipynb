Последовательность нуклеотидов хромосомы можно скачать по адресу https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/

```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/hg38/chromosomes/chr21.fa.gz
!gzip -d chr21.fa.gz
```

```python
%%time
!bowtie2-build chr21.fa chromosome_index
```

```python
!mkdir bowtie2_res
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF539PKS_trimmed.fastq \
          -S bowtie2_res/res_PKS.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF374VEK_trimmed.fastq \
          -S bowtie2_res/res_VEK.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF634CSE_trimmed.fastq \
          -S bowtie2_res/res_CSE.sam
```

## Peak calling

Помимо macs2 может быть использован пакет Homer для поиска пиков, а также macs3.

```python
!mkdir macs2_PKS
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_PKS.sam \
    -c bowtie2_res/res_CSE.sam \
 	  -f SAM \
	  --outdir macs2_PKS
```

```python
!mkdir macs2_VEK
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_VEK.sam \
    -c bowtie2_res/res_CSE.sam \
 	  -f SAM \
	  --outdir macs2_VEK
```
